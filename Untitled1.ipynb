```python
#Programa: Calculadora simples

def calculadora():
    print("=== Calculadora ===")
    a = float(input("Digite o primeiro número: "))
    b = float(input("Digite o segundo número: "))

    print("nEscolha a operação:")
    print("1. Soma")
    print("2. Subtração")
    print("3. Multiplicação")
    print("4. Divisão")

    op = input("Opção: ")
    if op == "1":
        print("Resultado: ", a + b)
    elif op == "2":
        print("Resultado:", a - b)
    elif op == "3":
      print("Resultado:", a * b)
    elif op == "4":
      print("Resultado:", a / b)
    else:
      print("Opção invalída!")

calculadora()
```
